```python
from nltk.corpus import stopwords
from nltk import word_tokenize
```

```python
import re

data = open("CISI.ALLnettoye","r").readlines()
data=[i[:len(i)-1] for i in data] #Enleve le '\n'

collection = {}
vocabulary={}
titles=[]
for line in range(len(data)):
    if data[line][:3]==".I ":
        doc=[]
        vocab=[]
    elif data[line-1][:3]==".I ":
        indice = int(data[line-1][3:])
        titre="Document_"+str(indice)+"-> "+str(data[line])
        titles.append(titre)
        vocab.append(data[line])
        vocabulary[indice] = vocab
    else:
        doc.append(data[line])
        collection[titre]=doc
        vocab.append(data[line])
        vocabulary[indice] = vocab

v = {}
for i in vocabulary.keys():
    s = []
    for j in vocabulary[i]:
        wt = word_tokenize(j)
        for w in wt:
            s.append(w.lower())
    v[i]=s
```

```python
from nltk.stem import WordNetLemmatizer

v_clean={}
wnl = WordNetLemmatizer()
for i in range(1,len(v)):
    liste=[]
    for word in v[i]:
        if len(word)>1 and word not in stopwords.words('english'):
            liste.append(word)
    v_clean[i]=liste
v_clean
```

```python
def TF(doc,maxi):
    # maxi : Nb de docs analysés (maximum = N)
    tfD={}
    for i in range(1,maxi):
        tfDict = {}
        bagOfWordsCount = len(doc[i].keys())
        for word, count in doc[i].items():
            tfDict[word] = float("{:.15f}".format(count / float(bagOfWordsCount)))
        tfD[i] = tfDict
    return tfD
```
